# Domain Analysis:-

* **symboling**- Cars are initially assigned with a risk factor symbol associated with its price (-3 denotes vehicle is pretty safe and +3 means the vehicle is not safe).
* **normalized_losses**- It is the relative average loss payment per insured vehicle year.
* **make**- It represents the manufacturing companies.
* **fuel type**- It relates to the type of fuel the car uses to drive. 
* **aspiration**-  Engines need air for combustion. The normally aspirated (std) and turbocharged and engines rely on the pressure inside the intake manifold to get air and fuel into the engine's cylinders.
* **num_of_doors**- Total number of doors present in a car
* **body**- It represents car category (Hatchback, Sedan, Convertible, MUV, SUV etc).
* **drive_wheels**-A drive wheel is a wheel of a motor vehicle that transmits force, transforming torque into tractive force from the tires to the road, causing the vehicle to move.
* **engine_location**- The powertrain layout of a motorised vehicle such as a car is often defined by the location of the engine and drive wheels. Layouts can roughly be divided into three categories: front-wheel drive (FWD), rear-wheel drive (RWD) and four-wheel drive (4WD).
* **wheel_base**- Wheelbase is the horizontal distance between the centers of the front and rear wheels.
* **length**- 1 in Target feature represents patient has liver disease and 2 represents patient has no liver disease.
* **width**-  width simply denotes the dimensions of the car measured cross-sectionally
* **height**-The height of the car is defined as the height measured from its lowest point (i.e. the point where the tyres touch the road) to the tallest point on the car’s roof
* **curb_weight**- It is the total mass of a vehicle with standard equipment and all necessary operating consumables, while not loaded with either passengers or cargo.
* **engine_type**- Engine type means a power source with defined characteristics such as fuel utilized, number of cylinders, displacement, and net brake horsepower.
* **num_of_cylinders**- The cylinder is the power unit of the engine. This is where fuel is burned and converted into mechanical energy that powers the vehicle. The number of cylinders in a typical car could be four, six or eight.
* **engine_size**- Engine size is the volume of fuel and air that can be pushed through a car’s cylinders and is measured in cubic centimetres (cc). Car engine sizes are normally specified in litres, which is rounded up to the nearest tenth of a litre. For example, a car that has a 1390cc engine would be described as a 1.4 litre.
* **fuel_system**- The fuel system is responsible for supplying fuel to the internal combustion engine. The fuel mixes together with air and then the spark plugs ignite the mixture to create the combustion power which moves the vehicle.
* **bore**- Bore is the diameter of each cylinder and stroke is the length that it travels when moving from bottom position to the top position.
* **stroke**- The stroke is the distance within the cylinder the piston travels.
* **compression_ratio**- It is defined as the ratio of the volume at bottom dead center and the volume at top dead center. Compression ratio determines the thermal efficiency of used thermodynamic cycle of the combustion engine.
* **horsepower**- Horsepower is a unit of measurement of power, or the rate at which work is done, usually in reference to the output of engines or motors.
* **peak_rpm**- The higher the RPM, the more the power output of the engine.
* **city_mpg**- The score a car will get on average in city conditions, with stopping and starting at lower speeds.
* **highway_mpg**- the average a car will get while driving on an open stretch of road without stopping or starting, typically at a higher speed.
* **price**- Price column represents the price of cars depending upon the features.

# K Nearest Neighbors(Regressor)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
## taking optimal k to determine how many nearest neighbors  to create

# create a list to store the error values for each k
error_rate = []

# Will take some time
for i in range(1,11):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test)) 
    # if predicted value is not equal to actual value, returns true (which is taken as 1) else false(0). 
    # Adds all the value and takes mean of it. So for each k-value, gets the mean of error.
    #print(np.mean(pred_i != y_test))
```

```python
# Lets plot the k-value and error rate

plt.figure(figsize=(10,6))
plt.plot(range(1,11),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# let's fit the data into KNN model and see how well it performs:
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(X_train,y_train)# Predict 
y_pred = knn.predict(X_test)
```
